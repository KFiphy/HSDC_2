```python
#Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

#Preprocessing
from sklearn.model_selection import train_test_split, cross_validate, GridSearchCV, ParameterGrid, RandomizedSearchCV, KFold,StratifiedKFold
from sklearn.model_selection import train_test_split

churn = pd.read_csv("/content/sample_data/Churn.csv")
churn.head()
```

```python
churn.isna().sum()
```

```python
churn.info()
```

```python
# Perform initial data preparation by converting the 'TotalCharges' column to numeric values and filling missing values with 0.
churn['TotalCharges'] = pd.to_numeric(churn['TotalCharges'], errors='coerce').fillna(0)
churn.info()
```

```python
#Convert the 'Churn' column to binary values, where 'No' is mapped to 0 and 'Yes' is mapped to 1.
# Convert 'Churn' to binary
churn['Churn'] = churn['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
# churn['SeniorCitizen'] = churn['SeniorCitizen'].astype(str)
churn['SeniorCitizen'] = churn['SeniorCitizen'].apply(lambda x: 'Yes' if x == 1 else 'No')
```

```python
churn.info()
```

```python
# Define numerical and categorical features
num_feat = ['tenure', 'MonthlyCharges', 'TotalCharges', ]
cat_feat = ['gender', 'SeniorCitizen' ,'Partner', 'Dependents', 'PhoneService', 'MultipleLines',
            'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
            'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod']
target = 'Churn'
```

```python
churn["SeniorCitizen"].value_counts()
```

```python
X = churn[num_feat + cat_feat]
y = churn[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Display the shapes of the train and test sets
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.compose import make_column_selector as selector
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# Feature engineering:

# The numerical features should be scaled using StandardScaler, convert the output back to a dataframe and put back the column names.

# The categorical features are one-hot encoded using OneHotEncoder(set sparse_output to false), convert the output back to a dataframe and put back the column names.

#  Combine scaled numerical and one-hot encoded categorical features into train and test set dataframes (use pd.concat)
```

```python
# Scale numerical features
scaler = StandardScaler()
X_train_num_scaled = scaler.fit_transform(X_train[num_feat])
X_test_num_scaled = scaler.transform(X_test[num_feat])

# Convert scaled numerical features back to a dataframe
X_train_num_scaled_df = pd.DataFrame(X_train_num_scaled, columns=num_feat)
X_test_num_scaled_df = pd.DataFrame(X_test_num_scaled, columns=num_feat)

# Display the first few rows of the scaled numerical features dataframe
print("Scaled Numerical Features (Train):")
print(X_train_num_scaled_df.head())
print("\nScaled Numerical Features (Test):")
print(X_test_num_scaled_df.head())
```

```python
# One-hot encode categorical features
encoder = OneHotEncoder(sparse=False)
X_train_cat_encoded = encoder.fit_transform(X_train[cat_feat])
X_test_cat_encoded = encoder.transform(X_test[cat_feat])

# # Get column names after one-hot encoding
encoded_columns = encoder.get_feature_names_out(cat_feat)

# Convert one-hot encoded categorical features back to a dataframe
X_train_cat_encoded_df = pd.DataFrame(X_train_cat_encoded, columns=encoded_columns)
X_test_cat_encoded_df = pd.DataFrame(X_test_cat_encoded,columns=encoded_columns)
```

```python
X_train_cat_encoded_df.head()
```

```python
# # Combine scaled numerical and one-hot encoded categorical features into train set dataframe
X_train = pd.concat([X_train_num_scaled_df, X_train_cat_encoded_df], axis=1)

# Combine scaled numerical and one-hot encoded categorical features into test set dataframe
X_test = pd.concat([X_test_num_scaled_df, X_test_cat_encoded_df], axis=1)

X_train.head()
```

```python
y_test.head()
```

```python
'''Use scikit learn to train a random forest and extra trees classifier, and use xgboost and lightgbm to train
an extreme boosting model and a light gradient boosting model.
Use random_state = 1 for training all models and evaluate on the test set.'''

# Initialize models with random_state=1
rf_model = RandomForestClassifier(random_state=1)
et_model = ExtraTreesClassifier(random_state=1)
xgb_model = XGBClassifier(random_state=1)
lgbm_model = LGBMClassifier(random_state=1)

# Train models & Predictions
rf_model.fit(X_train, y_train)
rf_preds = rf_model.predict(X_test)

et_model.fit(X_train, y_train)
et_preds = et_model.predict(X_test)

xgb_model.fit(X_train, y_train)
xgb_preds = xgb_model.predict(X_test)

lgbm_model.fit(X_train, y_train)
lgbm_preds = lgbm_model.predict(X_test)
```

```python
accuracy_score(y_test, rf_preds)
```

```python
accuracy_score(y_test, xgb_preds)
```

```python
accuracy_score(y_test, lgbm_preds)
```

```python
accuracy_score(y_test, et_preds)
```

```python
from sklearn.model_selection import RandomizedSearchCV

# Define the ExtraTreesClassifier estimator
et_model = ExtraTreesClassifier(random_state=1)

# Define the hyperparameter grid
n_estimators = [50, 100, 300, 500, 1000]
min_samples_split = [2, 3, 5, 7, 9]
min_samples_leaf = [1, 2, 4, 6, 8]
max_features = ['auto', 'sqrt', 'log2', None]
hyperparameter_grid = {
    'n_estimators': n_estimators,
    'min_samples_leaf': min_samples_leaf,
    'min_samples_split': min_samples_split,
    'max_features': max_features
}

# Initialize RandomizedSearchCV
random_search = RandomizedSearchCV(estimator=et_model, param_distributions=hyperparameter_grid,
                                   n_iter=10, scoring='accuracy', cv=5, n_jobs=-1, verbose=1, random_state=1)

# Perform RandomizedSearchCV
random_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = random_search.best_params_

# Print the best hyperparameters
print("Best hyperparameters:", best_params)
```

```python

# Define the ExtraTreesClassifier model with the best hyperparameters
best_et_model = ExtraTreesClassifier(random_state=1, **best_params)

# Train the model on the training data
best_et_model.fit(X_train, y_train)

# Predict on the test set
best_et_preds = best_et_model.predict(X_test)

# Calculate accuracy of the new optimal model
best_et_accuracy = accuracy_score(y_test, best_et_preds)

# Compare with initial ExtraTreesClassifier model accuracy
print( best_et_accuracy)
```

```python
# Get feature importances from the optimal ExtraTreesClassifier model
feature_importances = best_et_model.feature_importances_

# Create a DataFrame to display feature importances
feature_importance_df = pd.DataFrame({'Feature': X_train.columns, 'Importance': feature_importances})

# Sort the DataFrame by importance in descending order
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Display the top two most important features
top_two_features = feature_importance_df.head(2)
print("The two most important features:")
print(top_two_features)
```
